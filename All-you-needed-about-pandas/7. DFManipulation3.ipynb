```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/census.csv')
# And exclude state level summarizations, which have sum level value of 40
df = df[df['SUMLEV']==50]
```

## groupby

groupby的结果是在迭代器里面，可以for循环取出来

```python
for group, frame in df.groupby('STNAME'):
    # groupby() returns a tuple, where the first value is the
    # value of the key we were trying to group by, in this case 
    # a specific state name, and the second one is
    # projected dataframe that was found for that group
    
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + 
          ' have an average population of ' + str(avg))
```

groupby可以传入函数进行分组

```python
df = df.set_index('STNAME')
```

```python
df.head()
```

```python
def set_batch_number(item):
    if item[0] < 'M':
        return 0
    if item[0] < 'Q':
        return 1
    return 2

for group, frame in df.groupby(set_batch_number):
    print(f'{group}: {frame.shape}')
```

## transformation

transform 返回的数据大小与原本data的size相同，这使得数据的合并非常容易

```python
cols=['cancellation_policy','review_scores_value']
transform_df = df[cols].groupby(cols[0]).transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'review_scores_value':'mean_review_scores'}, axis='columns', inplace=True)
df=df.merge(transform_df, left_index=True, right_index=True)
df.head()
```

## filtering

```python
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9.2)
```

## apply

```python
df=pd.read_csv("datasets/listings.csv")
# And lets just include some of the columns we were interested in previously
df=df[['cancellation_policy','review_scores_value']]
df.head()
```

```python
def calc_mean_review_scores(group):
    # group is a dataframe just of whatever we have grouped by, e.g. cancellation policy, so we can treat
    # this as the complete dataframe
    avg=np.nanmean(group["review_scores_value"])
    # now broadcast our formula and create a new column
    group["review_scores_mean"]=np.abs(avg-group["review_scores_value"])
    return group

# Now just apply this to the groups
df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```

## Scales

我们可以将column数据类型更换为 category

构建category数据类型，并带上排序

```python
df=pd.DataFrame(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D'],
                index=['excellent', 'excellent', 'excellent', 'good', 'good', 'good', 
                       'ok', 'ok', 'ok', 'poor', 'poor'],
               columns=["Grades"])
```

```python
my_categories=pd.CategoricalDtype(categories=['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'], 
                           ordered=True)
# then we can just pass this to the astype() function
grades=df["Grades"].astype(my_categories)
grades.head()
```

可以看到，最终的类型，以及对应的排序

现在，我们就可以对value进行比较输出想要的rols

```python
grades[grades > 'C']
```

使用cut也可以将df数据分为制定数目的类别

```python
import numpy as np

df=pd.read_csv("datasets/census.csv")

df=df[df['SUMLEV']==50]

df=df.set_index('STNAME').groupby(level=0)['CENSUS2010POP'].agg(np.average)

df.head()
```

```python
pd.cut(df, 5)
```

可以看到，全部的数据分割成五部分。每个部分都有对应的区间

```python
S = pd.cut(df, 10)
group_df = pd.DataFrame(data = S).
```

```python
group_df.reset_index(inplace=True)
```

```python
group_df.set_index('CENSUS2010POP')
```

```python
group_df.drop('index',axis=1, inplace=True)
```

```python
group_df.set_index(group_df.columns[1], inplace=True)
```

```python
for group, states in group_df.groupby(level=0):
    
    print(f'{group}: {len(states)}')
```
